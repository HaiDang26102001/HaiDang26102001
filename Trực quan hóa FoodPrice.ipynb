```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/DD/Desktop/Data Analyst/FoodPrice_in_Turkey.csv')
df
```

# Mục tiêu 1: Vẽ biểu đồ cột so sánh giá gạo (Rice-Retail) tháng 12 năm 2019 của Ankara, Istanbul, Izmir và National Average.

```python
data1 = df[(df['Month'] == 12) & (df['Year'] == 2019) & (df['ProductName'] == 'Rice - Retail')]
data1
```

```python
plt.bar(data1['Place'],data1['Price'])
axs1 = plt.gca()
axs1.set_title('Rice Price in 12/2019',color='red')
axs1.set_xlabel('Place',color='red')
axs1.set_ylabel('Price',color='red')

plt.show()
```

=> Giá gạo tháng 12/2019 của Istanbu tương đương với Izmir, giá gạo tạo Ankara là thấp nhất

# Mục tiêu 3: Vẽ biểu đồ Scatter phân tích mối liên quan giữa giá gạo và giá gas trung bình quốc gia (National Average) tại Thổ Nhĩ Kì.

```python
df['ProductName'].unique()
```

```python
data3_x = df[(df['Place'] == "National Average") & (df['ProductName'] == 'Fuel (gas) - Retail') & (df['Year'] == 2019)]
data3_y = df[(df['Place'] == "National Average") & (df['ProductName'] == 'Rice - Retail') & (df['Year'] == 2019)]
```

```python
plt.scatter(data3_x['Price'],data3_y['Price'])
axs3 = plt.gca()
axs3.set_title('Relationship between Fuel Price and Rice Price',color='red')
axs3.set_xlabel('Gas',color='red')
axs3.set_ylabel('Rice',color='red')
plt.show()
```
