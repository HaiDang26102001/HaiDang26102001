```python
import turtle
import math
t = turtle.Turtle()
t.speed(0)

# Vẽ mặt trước ngôi nhà
t.color('Black','Blue')
t.begin_fill()
for i in range(2):
	t.fd(100)
	t.rt(90)
	t.fd(150)
	t.rt(90)
t.end_fill()

# Vẽ mái trước
t.color('Black','Pink')
t.begin_fill()
t.lt(60)
t.fd(100)
t.rt(120)
t.fd(100)
t.end_fill()

# Vẽ mái ngang
t.color('Black','Orange')
t.begin_fill()
t.lt(100)
t.fd(80)
t.lt(80)
t.fd(100)
t.lt(100)
t.fd(80)
t.lt(80)
t.fd(100)
t.end_fill()

# Vẽ tường bên nhà
t.color('Black','Yellow')
t.begin_fill()
t.rt(30)
t.fd(150)
t.lt(130)
t.fd(80)
t.lt(50)
t.fd(150)
t.end_fill()

# Vẽ mặt trời
t.color('Black','Yellow')

t.penup()
t.rt(30)
t.fd(210)
t.pendown()

t.rt(60)
t.begin_fill()
for i in range(4):

    for i in range(45):
        t.fd(1)
        t.lt(1)
    t.rt(90)
    t.fd(30)
    t.lt(180)
    t.fd(30)
    t.rt(90)

    for i in range(45):
        t.fd(1)
        t.lt(1)
    t.rt(90)
    t.fd(60)
    t.lt(180)
    t.fd(60)
    t.rt(90)
t.end_fill()

# Quay lại nhà để thêm cửa chính và cửa sổ
t.rt(120)
t.penup()
t.fd(250)
t.lt(30)
t.pendown()

# Vẽ cửa sổ
t.color('Black','Brown')
t.begin_fill()
t.fd(50)
t.rt(50)
t.fd(33.922)
t.rt(130)
t.fd(50)
t.rt(50)
t.fd(33.922)
t.end_fill()

# Vẽ cửa chính
t.color('Black','Green')
t.rt(180)
t.penup()
t.fd(100)
t.lt(50)
t.pendown()
t.begin_fill()
t.fd(103)
t.rt(90)
t.fd(40)
t.rt(90)
t.fd(103)
t.rt(90)
t.fd(40)
t.end_fill()

# Vẽ cây
t.lt(180)
t.penup()
t.fd(300)
t.pendown()
t.color('Black','Brown')
t.begin_fill()
for i in range(2):
    t.lt(90)
    t.fd(70)
    t.lt(90)
    t.fd(25)
t.end_fill()

t.color('Black','Green')
t.begin_fill()

t.fd(32.5)
t.rt(120)
t.fd(90)
t.rt(120)
t.fd(90)
t.rt(120)

t.fd(45)
t.rt(90)
t.end_fill()
t.penup()
t.fd(45*math.sqrt(3))

t.pendown()
t.color('Black','Green')
t.begin_fill()
t.lt(90)
t.fd(40)
t.rt(120)
t.fd(80)
t.rt(120)
t.fd(80)
t.rt(120)
t.fd(40)
t.rt(90)
t.end_fill()
t.penup()
t.fd(40*math.sqrt(3))

t.pendown()
t.color('Black','Green')
t.begin_fill()
t.lt(90)
t.fd(35)
t.rt(120)
t.fd(70)
t.rt(120)
t.fd(70)
t.rt(120)
t.fd(35)
t.end_fill()
```
