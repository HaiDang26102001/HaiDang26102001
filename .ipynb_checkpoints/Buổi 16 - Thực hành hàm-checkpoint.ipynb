### Bài 2

Viết hàm `xoachuoi` có dữ liệu đầu vào là 2 chuỗi a và b, hãy tìm trong chuỗi a có chuỗi b hay không, nếu có hãy xóa toán bộ chuỗi b trong a.

Sau đó trả về chuỗi a mới

Ví dụ:

chuoi1 = "xin chào các bạn, các bạn có khỏe không?"

chuoi2 = "các bạn"

chuoi3 = xoachuoi(chuoi1,chuoi2)

print(chuoi3)

Kết quả in ra là:
xin chào ,  có khỏe không

```python
a = input('Chuỗi 1: ')
b = input('Chuỗi 2: ')
def xoachuoi(x,y):
    for i in range(x.count(y)):
        x = x[:x.find(y)] + x[x.find(y) + len(y) + 1:]
    return x
print(xoachuoi(a,b))
```

### Bài 3

Viết hàm `kiemtramatkhau` với dữ liệu đầu vào là 1 chuỗi có công dụng như sau:

Kiểm tra xem chuỗi đầu vào có phải là mật khẩu mạnh hay không, nếu có trả về True, nếu không trả về False

Mật khẩu mạnh là mật khẩu phải có ít nhất 1 chữ hoa, 1 chữ thường, 1 số và 1 ký tự đặc biệt

```python
matkhau = input('Nhập mật khẩu: ')

def ktramk(a):
    dem = 0
    for i in a:
        if i.isupper() == True:
            dem += 1
            break

    for i in a:
        if i.islower() == True:
            dem += 1
            break

    for i in a:
        if i.isnumeric() == True:
            dem += 1
            break

    for i in a:
        if not i.isupper() == True and not i.islower() == True and not i.isnumeric() == True:
            dem += 1
            break

    if dem == 4:
        return True
    else:
        return False

print(ktramk(matkhau))
        
        
        
            
```

### Bài 4

Viết hàm có 5 dữ liệu đầu vào là 5 số nguyên, tìm giá trị lớn nhì trong 5 số đó

```python
so1 = float(input('Nhập số thứ 1: '))
so2 = float(input('Nhập số thứ 2: '))
so3 = float(input('Nhập số thứ 3: '))
so4 = float(input('Nhập số thứ 4: '))
so5 = float(input('Nhập số thứ 5: '))


def monday(a,b,c,d,e):
    max1 = max(a,b)
    max2 = min(a,b)
    if c > max1:
        max2 = max1
        max1 = c
    elif c > max2 and c != max1:
        max2 = c
        
    if d > max1:
        max2 = max1
        max1 = d
    elif d > max2 and d != max1:
        max2 = d 
    
    if e > max1:
        max2 = max1
        max1 = e
    elif e > max2 and e != max1:
        max2 = e
    
    return max2
        
print(monday(so1,so2,so3,so4,so5))      
 


    
        
            
    
```

### Bài 5

Nhập vào 2 cặp tọa độ tâm đường tròn và 2 bán kính R

Hãy vẽ 2 hình tròn theo tọa độ và bán kính trên (tọa độ thứ nhất và bán kính thứ nhất là hình tròn thứ nhất, tọa độ thứ 2 với bán kính thứ 2 là hình tròn thứ hai)

Viết hàm kiểm tra xem 2 hình tròn trên cắt nhau bao nhiêu điểm

```python
import math
import turtle 
t = turtle.Turtle()

x1= float(input('Nhập tọa độ x hình tròn 1: '))
y1= float(input('Nhập tọa độ y hình tròn 1: '))
x2= float(input('Nhập tọa độ x hình tròn 2: '))
y2= float(input('Nhập tọa độ y hình tròn 2: '))

r1 = int(input('Nhập bán kính ht 1: '))
r2 = int(input('Nhập bán kính ht 2: '))

t.penup()
t.goto(x1,y1)
t.pendown()
t.circle(r1)

t.penup()
t.goto(x2,y2)
t.pendown()
t.circle(r2)

R = r1 + r2
kct = math.sqrt((x2-x1)**2 - (y2-y1)**2)

# Tìm r: bán kính của hình tròn lớn hơn
# Tìm bkn: bán kính của hình tròn nhỏ hơn
if r1 > r2:
    r = r1
    bkn = r2
elif r2 > r1:
    r = r2
    bkn = r1

def check():
    giaodiem = 0
    '''
    TH giao điểm = 0 có 2 khả năng xảy ra:
    - 1: KC 2 tâm lớn hơn tổng 2 bán kính
    - 2: Hình tròn nhỏ nằm trọn trong hình tròn lớn và không cắt hình tròn lớn
    '''
    if kct > R or (kct < r and bkn < (r - kct)):
        giaodiem = 0
    
    elif kct == 0 and r1 == r2:
        giaodiem = 'Vô số'
    
    elif kct < R:
        giaodiem = 2
    '''
    TH giao điểm = 1 có 2 khả năng xảy ra:
    - 1: KC 2 tâm = tổng 2 bán kính
    - 2: Hình tròn nhỏ nằm trong hình tròn lớn và cắt hình tròn lớn tại 1 điểm
    '''
    elif kct == R or (kct < r and bkn == (r - kct)):
        giaodiem = 1
    
    return giaodiem
    


print('Số giao điểm của 2 đường tròn là :',check())
    



```

### Bài 6

Nhập vào 2 tọa độ là 2 đỉnh của một hình chữ nhật

Sau đó

Nhập vào tọa độ tâm đường tròn và bán kính R

Hãy vẽ hình hình chữ nhật và đường tròn trên

Viết hàm kiểm tra xem đường tròn trên và hình chữ nhật cắt nhau tại bao nhiêu điểm

```python
import math
import turtle
t = turtle.Turtle()
hcnX1 = float(input('Nhập tọa độ x đỉnh 1 của hcn: ')) # gọi là đỉnh A
hcnY1 = float(input('Nhập tọa độ y đỉnh 1 của hcn: '))

hcnX2 = float(input('Nhập tọa độ x đỉnh 2 của hcn: ')) # Gọi là đỉnh B
hcnY2 = float(input('Nhập tọa độ y đỉnh 2 của hcn: '))

htX = float(input('Nhập tọa độ x tâm đường tròn: '))
htY = float(input('Nhập tọa độ y tâm đường tròn: '))

R = float(input('Nhập bán kính đường tròn: '))

cd = float(input('Nhập độ dài cạnh còn lại của hình chữ nhật: '))

# viết 1 hàm tính khoảng cách giữa 2 toạ độ bất kì
def kc(x1,y1,x2,y2):
    d = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    return d

### Tìm toạ độ 2 đỉnh còn lại của hcn

a = -1/((hcnY1 - hcnY2)/(hcnX1-hcnX2))
# lập pt đường thẳng đi qua AC và BD

b1 = hcnY1 - a*hcnX1
b2 = hcnY2 - a*hcnX2


# viết hàm giải pt bậc 2 để tìm toạ độ x của 2 đỉnh còn lại
def ptbac2(a,b,c):
    x = 0
    if a == 0:
        x = -c/b
    else:
        delta = b*b - 4*a*c
        if delta < 0:
            print('Pt vô nghiệm')
        elif delta == 0:
            x = -b/(2*a)
        else:
            x = (-b + math.sqrt(delta))/(2*a)
    
    return x 

# Tính ra được toạ độ của 2 đỉnh còn lại
hcnX3 = ptbac2(a**2+1, 2*a*(b1-hcnY1)-2*hcnX1, hcnX1**2 + (b1 - hcnY1)**2 - cd**2)
hcnY3 = a*hcnX3 + b1
hcnX4 = ptbac2(a**2+1, 2*a*(b2-hcnY2)-2*hcnX2, hcnX2**2 + (b2 - hcnY2)**2 - cd**2)
hcnY4 = a*hcnX4 + b2

# Vẽ hình chữ nhật
t.penup()
t.goto(hcnX1,hcnY1)
t.pendown()
t.goto(hcnX2,hcnY2)
t.goto(hcnX4,hcnY4) # gọi là đỉnh C
t.goto(hcnX3,hcnY3) # Gọi là đỉnh D
t.goto(hcnX1,hcnY1)

# Vẽ hình tròn
t.penup()
t.goto(htX,htY)
t.bk(R)
t.rt(90)
t.pendown()
t.circle(R)

# Tìm điểm trên hình chữ nhật mà gần nhất với tâm đường tròn

# Viết 1 hàm kiểm tra 
def ktra(a,b = R):
    if a == b:
        return 1
    elif a > b:
        return 0
    else:
        return 2


# So sánh khoảng cách từ tâm đường tròn tới 8 điểm trên hình chữ nhật

'''
            P
  __|_________._________|__
  D |                   | C
    |                   |
  Q .                   . N
    |                   |
  __|_________._________|__
  A |        M          | B


  Thuật toán sẽ là: - Tính khoảng cách từ tâm đường tròn tới 8 điểm A,B,C,D,M,N,P,Q để tìm ra kc nào là nhỏ nhất
                    - Sau đó so sánh khoảng cách đó với R của hình tròn để tìm ra số giao điểm.
'''

# Toạ độ 4 điểm M,N,P,Q lần lượt là:
hcnXm = (hcnX1 + hcnX2)/2
hcnYm = (hcnY1 + hcnY2)/2

hcnXn = (hcnX2 + hcnX4)/2
hcnYn = (hcnY2 + hcnY4)/2

hcnXp = (hcnX4 + hcnX3)/2
hcnYp = (hcnY4 + hcnY3)/2

hcnXq = (hcnX1 + hcnX3)/2
hcnYq = (hcnY1 + hcnY3)/2

# Tìm kcnn
kcnn = min(kc(htX,htY,hcnX1,hcnY1), kc(htX,htY,hcnX2,hcnY2), kc(htX,htY,hcnX3,hcnY3), kc(htX,htY,hcnX4,hcnY4), kc(htX,htY,hcnXm,hcnYm), kc(htX,htY,hcnXn,hcnYn), kc(htX,htY,hcnXp,hcnYp), kc(htX,htY,hcnXq,hcnYq))

# Viết hàm kiểm tra giao điểm
def giaodiem():
    sogiaodiem = 0
    if kcnn == R:
        sogiaodiem = 1
    elif kcnn > R:
        sogiaodiem = 0
    else:
        sogiaodiem = 2
    return sogiaodiem

print('Số giao điểm của hình chữ nhật và hình tròn là: ',giaodiem())
```

### Bài 7

Viết hàm kiểm tra số nguyên tố

Viết hàm kiểm tra số hoàn hảo

Viết hàm kiểm tra số chính phương

Viết hàm kiểm tra số Armstrong

Nhập vào một chuỗi, hãy lọc và tìm ra những số ở trong chuỗi (2 chữ số liền nhau là một chuỗi)

Ví dụ chuỗi: "AB25SDc5DSjjh52"

Trong chuỗi này ta có 3 số: 25, 5 và 52

Sau khi lọc số, dùng các hàm trên để kiểm tra và thực hiện như sau:

Cho biến S

- Nếu số đó là số nguyên tố, ta thực hiện S += số đó
- Nếu số đó là số hoàn hảo, ta thực hiện S -= số đó
- Nếu số đó là số chính phương, ta thực hiện S /= số đó
- Nếu đó là số Armstrong ta cho S = 0

Hãy tính và in S

```python
import math
def snt(a):
    a = int(a)
    dem = 0
    for i in range(1,a):
        if a % i == 0:
            dem += 1
    if dem == 1:
        return True
    return False

def shh(b):
    b = int(b)
    S = 0
    for i in range(1,b):
        if b % i == 0:
            S += i
    if S == b:
        return True
    return False

def scp(c):
    c = int(c)
    if int(math.sqrt(c)) == math.sqrt(c):
        return True
    return False

def sas(d):
    S = 0
    for i in range(len(str(d))):
        S += int(int(d[i])**len(str(d)))
    if S == d:
        return True
    return False

chuoi = input('Nhập chuỗi: ')
chuoi += ' '
bientam = ''
S = 0
for k in chuoi:
    if k.isnumeric():
        bientam += k
    else:
        if bientam != '':
            print(bientam)
            if snt(bientam) == True:
                S += int(bientam)
            elif shh(bientam) == True:
                S -= int(bientam)
            elif scp(bientam) == True:
                S /= int(bientam) 
            elif sas(bientam) == True:
                S = 0
            bientam = ''

print('S =',S)
    
```
