```python
D = {1:'Phở bò',2:'Phở gà'}
def Hien_thi_san_pham(D):
    # Hàm hiển thị H - hiển thị sản phẩm
    print('--------------------')
    print('\n')
    for i in D:
        print(str(i)+':',D[i])
    print('\n')
    print('--------------------')


def Them_san_pham(D):
    # Hàm T - thêm sản phẩm
    while True:
        ID = int(input('Nhập vào ID sản phẩm muốn thêm: '))
        if ID == 0:
            break
        elif ID in D:
            print('ID đã tồn tại, vui lòng nhập ID khác')
        elif ID not in D:
            product = input('Nhập sản phẩm muốn thêm: ')
            D[ID] = product
            
def Sua_san_pham(D):
    while True:
        ID = int(input('Nhập vào ID sản phẩm muốn sửa: '))
        if ID == 0:
            break 
        elif ID not in D:
            print('ID không tồn tại, vui lòng nhập lại ID')
        elif ID in D:
            product = input('Nhập tên sản phẩm muốn sửa: ')
            D[ID] = product

def Xoa_san_pham(D):
    while True:
        ID = int(input('Nhập ID sản phẩm muốn xoá: '))
        if ID == 0:
            break
        elif ID not in D:
            print('ID không tồn tại, vui lòng nhập lại ID')
        elif ID in D:
            P = input('Bạn có chắc muốn xoá sản phẩm này? Nếu có gõ C, nếu không gõ bất kì: ')
            if P == 'C':
                del D[ID]
            else:
                    break


print('Nhập H - hiển thị sản phẩm\nNhập T - thêm sản phẩm\nNhập S - sửa sản phẩm\nNhập X - xoá sản phẩm\nNhập 0 để thoát chương trình')


while True:
    Y = input('Nhập chức năng yêu cầu: ')

    if Y == '0':
        print('Cảm ơn bạn đã sử dụng!')
        break     
    elif Y == 'H':
        Hien_thi_san_pham(D)
    elif Y == 'T':
        Them_san_pham(D)
    elif Y == 'S':
        Sua_san_pham(D)
    elif Y == 'X':
        Xoa_san_pham(D)
    else:
        print('Không có chức năng này, vui lòng nhập lại')
```
