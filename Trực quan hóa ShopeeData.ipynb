```python
df['response_time'] = pd.to_datetime(df['response_time'],format=' %H:%M:%S')
```

```python
df['response_time_second'] = (df['response_time'].dt.hour*60 + df['response_time'].dt.minute)*60 + df['response_time'].dt.second
df
```

```python
c3 = df.loc[:,['rating_bad','response_time_second']]
c3.sort_values(by='response_time_second',inplace=True)
c3
```

```python
plt.plot(c3['response_time_second'],c3['rating_bad'])
plt.show()
```
